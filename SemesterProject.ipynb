## Why gather all these stats? What is eFG%? What is Pace?

To understand whether the NBA has truly declined, we need more than just viewership numbers.  
We need to look at how the game itself has changed over time and whether those changes are for better or worse.

**Effective Field Goal Percentage (eFG%)**: a more accurate way of measuring shooting efficiency than basic field goal percentage.
It accounts for the fact that three-point shots are worth more than two-point shots, and gives us a better sense of how well a team is scoring relative to its shot selection.

 **Pace**: refers to the number of possessions a team uses per game.
 
 A faster pace means more possessions, more shot attempts, and often a more aggressive style of play.

The bar chart below shows the average eFG% for each NBA Finals team from 1981 to 2024.  
By comparing these numbers across decades, we can see whether teams are becoming more efficient and how the game has evolved in terms of offensive quality. 


```python
import matplotlib.pyplot as plt
import pandas as pd

team_efg_averages = {
    "1981 Celtics": team_avg_efg_1981_celtics,
    "1981 Rockets": team_avg_efg_1981_rockets,
    "1998 Bulls": team_avg_efg_1998_bulls,
    "1998 Jazz": team_avg_efg_1998_jazz,
    "2008 Celtics": team_avg_efg_2008_celtics,
    "2008 Lakers": team_avg_efg_2008_lakers,
    "2016 Cavaliers": team_avg_efg_2016_cavs,
    "2016 Warriors": team_avg_efg_2016_warriors,
    "2024 Mavericks": team_avg_efg_2024_mavs,
    "2024 Celtics": team_avg_efg_2024_celtics
}

efg_series = pd.Series(team_efg_averages).sort_index()

plt.figure(figsize=(10, 6))
efg_series.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average eFG% by NBA Finals Team (1981–2024)")
plt.ylabel("Effective Field Goal Percentage")
plt.xlabel("Team")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

def calculate_pace(df):
    possessions = df['FGA'].sum() + 0.44 * df['FTA'].sum() + df['TOV'].sum() - df['ORB'].sum()
    total_minutes = df['MP'].sum()
    return 48 * possessions / total_minutes

def calculate_efg(fgm, fg3m, fga):
    return (fgm + 0.5 * fg3m) / fga

finals_data = {
    "1981 Celtics": celtics_1981,
    "1981 Rockets": rockets_1981,
    "1998 Bulls": bulls_1998,
    "1998 Jazz": jazz_1998,
    "2008 Celtics": celtics_2008,
    "2008 Lakers": lakers_2008,
    "2016 Cavaliers": cavs_2016,
    "2016 Warriors": warriors_2016,
    "2024 Mavericks": mavs_2024,
    "2024 Celtics": celtics_2024
}

results = []
for team, df in finals_data.items():
    pace = calculate_pace(df)
    efg = calculate_efg(df['FG'].sum(), df['3P'].sum(), df['FGA'].sum())
    results.append({"Team": team, "Pace": pace, "eFG%": efg})

pace_efg_df = pd.DataFrame(results).sort_values("Team").reset_index(drop=True)
print(pace_efg_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(pace_efg_df["Pace"], pace_efg_df["eFG%"], color="darkorange", edgecolor="black")
for i in range(len(pace_efg_df)):
    plt.text(pace_efg_df["Pace"][i], pace_efg_df["eFG%"][i], pace_efg_df["Team"][i], fontsize=8)

plt.title("Pace vs Effective Field Goal % (NBA Finals Teams)")
plt.xlabel("Pace (Estimated Possessions per 48 Min)")
plt.ylabel("Effective Field Goal Percentage")
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

## Why Do These Stats Matter?
 
My initial hope was to compare the number of 3-point attempts to overall field goal attempts,expecting to find that the percentage of 3 point shots taken has always been consistent. However, this wasn't the case but after more thinking the data presented two trends:

- Effective Field Goal Percentage (eFG%) has steadily increased, showing that teams are shooting more efficiently.
- Pace of play has fluctuated, but modern Finals teams tend to combine solid pace with much better shot efficiency.

## This tells us something important:

Even if teams aren't playing significantly faster, they're playing smarter. The average player today is more skilled(aka) taking better shots and making them at a higher rate.


So with skills all across the league on the rise, its not the style of play thats lacking so what would make the viewership go down?



## If the game is better, why is viewership down?

We’ve shown that today's NBA players are more efficient and skilled than ever before.  
So if it's not a drop in quality, what *is* driving the decline in traditional TV numbers?

To answer that, we have to zoom out.

For the **SEVENTH** consecutive year, the MVP award has been claimed by a player born outside the United States.​

The NBA isn’t just an American product anymore, it’s a **global phenomenon**.


While **cable viewership in the U.S. is shrinking**, the league is **booming internationally** and on **digital platforms**.  
YouTube views, social media engagement, app usage, and international fanbases are growing at unprecedented rates.



This next chart puts that contrast in perspective:


```python
df_trends = pd.read_csv("https://raw.githubusercontent.com/CassiusG13/NBA-Viewership-Analysis-/main/ViewingTrends.csv")

df_trends['Value'] = df_trends['Value'].replace('%', '', regex=True).astype(float)

plt.figure(figsize=(10, 6))
bars = plt.barh(df_trends['Metric'], df_trends['Value'], color='teal')

for bar, metric in zip(bars, df_trends['Metric']):
    if "Cable" in metric or "TV" in metric:
        bar.set_color('salmon')  # decline
    else:
        bar.set_color('mediumseagreen')  # growth

# Labeling
plt.title('Global Growth vs. Cable Decline: How Fans Watch the NBA (2024)')
plt.xlabel('Value (%) or Millions')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## So... is the NBA actually in decline?

Absolutely not.

What we’re seeing is a **shift in how basketball is being consumed** not a drop in interest.

Cable TV is no longer the main platform for sports fans, especially younger audiences and international viewers.  
Meanwhile, the NBA is thriving where those fans actually are:  
on **YouTube**, **TikTok**, **Instagram**, the **NBA App**, and all across **Europe, Asia, and Africa**.

The data tells a clear story:
**Traditional metrics like TV ratings can no longer capture the true scale of NBA popularity.**

## Bummer! You don’t have cable. You can’t rely on League Pass. So... how can you actually watch a game?

For many fans — especially younger ones — the answer has become simple:  
**You stream it illegally.**

Whether it’s through Reddit threads, Streameast links, or private Discord servers,  
pirated streams have quietly become one of the most common ways people watch NBA games.

And the numbers back it up.

The chart below breaks down **how common illegal sports streaming is by age group**.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load piracy data from GitHub
piracy_df = pd.read_csv("https://raw.githubusercontent.com/CassiusG13/NBA-Viewership-Analysis-/main/PiracyAges.csv")

# Plot as a bar chart
plt.figure(figsize=(6, 4))
plt.bar(piracy_df['Age Group'], piracy_df['% Watched Pirated Sports'], color='tomato', edgecolor='black')

plt.title("Percentage of Fans Watching Pirated Sports by Age Group")
plt.ylabel("Percent Pirating")
plt.xlabel("Age Group")
plt.ylim(0, 60)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## What This Chart Tells Us

The younger you are, the more likely it is that you’re **not** watching the NBA through traditional methods.

- **Over half (54%) of adults aged 18–24** admit to streaming sports content illegally.
- That number drops to **30% for ages 35–55**, and even further to **just 13% for those over 55**.

This isn’t about ethics, it’s about **access**.

Younger fans aren’t tuning out because they don’t like basketball.  
They’re tuning out of **cable** and the systems that restrict how they can watch the games they love.

And here’s where age becomes even more important:  
**Younger fans are more tech savvy.** They know how to navigate the internet, find alternative streaming sites, and use platforms to find games.

When nearly every legal option is either paywalled or blocked, **piracy becomes the default** especially for fans raised on the internet.


## Discussion

The data tells a very different story than the headlines. While traditional TV ratings have declined, this doesn't mean people have stopped watching the NBA , they’ve simply changed *how* they watch.

Younger fans are overwhelmingly turning to streaming, social media, and even piracy not because they aren’t interested, but because cable and League Pass don’t meet their expectations. In fact, barriers like local blackouts and rising subscription costs often push fans away from legitimate options altogether.

At the same time, the NBA's global footprint is expanding faster than ever. International MVPs, billions of views on social platforms, and exploding interest in Europe, Asia, and Africa all point to a league that is *more relevant than ever* just in new ways.

So with all of this evidence how could anyone say that the state of Basketball is in bad shape?
